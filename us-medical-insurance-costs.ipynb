# U.S. Medical Insurance Costs

Portfolio project on U.S. Medical Insurance Costs. After organizing the data, I will use my coding skills to provide basic, descriptive statistics of the data. After that I will try and look into the factors that impact the annual medical charges the most.

Let's start with importing the CSV module, and the pandas module for analysis.

```python
# importing pandas and csv modules
import pandas as pd
import csv

# variables that will contain the counts for both males and females.
count_male = 0
count_female = 0

with open('insurance.csv') as insurance_data:
    reader = csv.reader(insurance_data)
    for row in reader:
        if row[1] == 'male':
            count_male += 1
        else:
            count_female += 1
            
    print(count_male, count_female, sep='\n')
    print(f'The total number of people in this sample is {(count_male + count_female) - 1}.')
```

```python
smoker_count = 0
nonsmoker_count = 0
sum_of_smokers = 0
sum_of_nonsmokers = 0
with open('insurance.csv') as insurance_data:
    headings = insurance_data.readline().strip('\n').strip(',')
    reader = csv.reader(insurance_data)
    for row in reader:
        if row[4]== 'yes':
            smoker_count += 1
            sum_of_smokers += float(row[6])
        else:
            nonsmoker_count += 1
            sum_of_nonsmokers += float(row[6])
            
    avg_cost_of_smokers = sum_of_smokers / smoker_count
    avg_cost_of_nonsmokers = sum_of_nonsmokers / nonsmoker_count
    print(avg_cost_of_smokers, avg_cost_of_nonsmokers, sep=',')
    
    increase_if_smoker = avg_cost_of_smokers - avg_cost_of_nonsmokers
    percentage = increase_if_smoker / avg_cost_of_nonsmokers * 100
    print(percentage)
```

```python
df = pd.read_csv('insurance.csv')
print(df)
```

```python
is_smoker_under_35 = 0
is_nonsmoker_under_35 = 0
is_smoker_above_35 = 0
is_nonsmoker_above_35 = 0

for index, row in df.iterrows():
    age = row['age']
    smoker = row['smoker']
    
    if age <= 35 and smoker == 'yes':
        is_smoker_under_35 += 1
    else:
        is_nonsmoker_under_35 += 1
    if age > 35 and smoker == 'yes':
        is_smoker_above_35 += 1
    else:
        is_nonsmoker_above_35 += 1
        
print('Number of smokers who is over 35 years is ', is_smoker_under_35)
print('Number of smokers who is under 35 years is ', is_smoker_above_35)

smoker_under_35_percent =  round(is_smoker_under_35 * 100 / smoker_count,2)
smoker_above_35_percent =  round(is_smoker_above_35 * 100 / smoker_count,2)

print("Smokers under 35 percent", smoker_under_35_percent)
print("Smokers above 35 percent", smoker_above_35_percent)
```
